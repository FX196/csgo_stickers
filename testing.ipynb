```python
raw_features = pd.read_csv("features.csv")
raw_features = raw_features.applymap(lambda s:s.lower() if type(s) == str else s)
raw_features
```

```python
from sklearn import preprocessing

def prep(df, col):
    le = preprocessing.LabelEncoder()
    if col == 'tournament':
        lst = [x.lower() for x in list(tournaments.keys())]
        le.fit(lst)
    elif col == 'team':
        lst = [x.lower() for x in list(teams.keys())]
        le.fit(lst)
    elif col == 'country':
        lst = [x.lower() for x in list(teams.values())]
        le.fit(lst)
    
    return le

le_tour = prep(raw_features, 'tournament')
le_team = prep(raw_features, 'team')
le_coun = prep(raw_features, 'country')

#le_tour.transform(x['tournament'])
```

```python
tour_list = le_tour.transform(raw_features['tournament'])
se = pd.Series(tour_list)
raw_features['tournament'] = se.values

team_list = le_team.transform(raw_features['team'])
se = pd.Series(team_list)
raw_features['team'] = se.values

coun_list = le_coun.transform(raw_features['country'])
se = pd.Series(coun_list)
raw_features['country'] = se.values

raw_features
```

```python
from sklearn.model_selection import train_test_split
def x_y_split(data):
    x = ['tournament','team','placing','country']
    y = ['value']
    return data[x], data[y]
x,y = x_y_split(raw_features)
```

```python
def tour_split(data, tour):
    """
    tour : number from 0-9
    returns: everything else, specific tour
    """
    a = data.loc[data['tournament'] != tour]
    b = data.loc[data['tournament'] == tour]
    return a,b
```

```python
train, test = tour_split(raw_features, 1)
train_x, train_y = x_y_split(train)
test_x, test_y = x_y_split(test)
```

```python
xgreg = xgb.XGBRegressor(seed = 10, random_state = 10)
xgreg.fit(train_x,train_y)
xgreg.predict(test_x)
```

```python
def cleanup(data):
    col = 'Date'
    price = 'Price'
    for i in len(data)-1:
        if data[col][i] == data[col][i+1]:
            j = i+1
            count = 2
            val = data[price][i] + data[price][i+1]
            while j < len(data):
                if data[col][j] == data[col][j+1]:
                    count += 1
                    val += data[price][j+1]
                else:
                    break
            new = val / count
            data.iloc[i][1] = new
            data.drop(data.index[[i+1,j]])
            
    return data
```
